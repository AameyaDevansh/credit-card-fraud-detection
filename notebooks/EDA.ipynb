# üß™ Exploratory Data Analysis: Credit Card Fraud Detection

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

%matplotlib inline

# Load dataset
df = pd.read_csv('../data/creditcard.csv')
df.head()
```

## üßæ Basic Information

```python
df.info()
df.describe()
df.isnull().sum()
```

## ‚öñÔ∏è Class Distribution

```python
sns.countplot(x='Class', data=df)
plt.title("Class Distribution (0 = Normal, 1 = Fraud)")
plt.show()

fraud_percent = df['Class'].mean() * 100
print(f"Fraudulent transactions: {fraud_percent:.4f}%")
```

## üí∞ Feature Distributions: Amount & Time

```python
plt.figure(figsize=(10,5))
sns.histplot(df['Amount'], bins=100, kde=True)
plt.title("Distribution of Transaction Amount")
plt.show()

plt.figure(figsize=(10,5))
sns.histplot(df['Time'], bins=100, kde=True)
plt.title("Distribution of Time (in seconds)")
plt.show()
```

## üîó Correlation Matrix

```python
plt.figure(figsize=(15,10))
corr_matrix = df.corr()
sns.heatmap(corr_matrix, cmap='coolwarm', linewidths=0.5)
plt.title("Correlation Heatmap")
plt.show()
```

## üîç Comparing Feature Distributions by Class

```python
features = ['V1', 'V2', 'V3', 'V4']
for col in features:
    plt.figure(figsize=(10,4))
    sns.kdeplot(df[df['Class']==0][col], label="Non-Fraud", fill=True)
    sns.kdeplot(df[df['Class']==1][col], label="Fraud", fill=True)
    plt.title(f"Distribution of {col} by Class")
    plt.legend()
    plt.show()
```

## üìå EDA Insights
- Dataset is highly imbalanced (~0.17% fraud).
- No missing values found.
- Amount and Time need to be scaled.
- Some PCA components show clear distribution differences between classes.
- These insights will guide our preprocessing and model choices.
